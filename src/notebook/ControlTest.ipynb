```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import control as cs
import control.pzmap as pz
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed
```

# P1 Element

Erstellung des P1 Elements mit variablen K und T.

```python
K = 1.0
T = 1.0

P1 = cs.tf([K],[T,1])
```

Ãœbertragungsfunktion im s-Space

```python
print(P1)
```

Ausgabe der Sprungantwort vom P1 Element.

```python
y,t = cs.step(P1)
plt.plot(t,y)
plt.xlim(0,3)
plt.xlabel("Zeit in s")
```

Analyse der Pol- Nullstellen.

```python
pole, zero = pz.pzmap(P1)
```

Bodediagram vom P1 Element.

```python
mag, phase, omega = cs.bode(P1,dB=True,Hz=True)
```

Nyquistdiagram vom P1 Element.

```python
real, imag, fraq = cs.nyquist(P1)
```
